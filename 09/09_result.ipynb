# Logistic Regression based on feature functions 

## import library

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import time                                                
import util_solution as util
```

## load training data

```python
fname_data1 = '09_data1.csv'
fname_data2 = '09_data2.csv'

data1 = np.genfromtxt(fname_data1, delimiter=',')
data2 = np.genfromtxt(fname_data2, delimiter=',')

num_data1 = data1.shape[0]
num_data2 = data2.shape[0]

x1      = np.zeros(num_data1)
y1      = np.zeros(num_data1)
label1  = np.zeros(num_data1)

x2      = np.zeros(num_data2)
y2      = np.zeros(num_data2)
label2  = np.zeros(num_data2)

for i in range(num_data1):
    x1[i]       = data1[i,0]
    y1[i]       = data1[i,1]
    label1[i]   = data1[i,2]

for i in range(num_data2):
    x2[i]       = data2[i,0]
    y2[i]       = data2[i,1]
    label2[i]   = data2[i,2]

x1 = np.reshape(x1, (num_data1, 1))
y1 = np.reshape(y1, (num_data1, 1))
label1 = np.reshape(label1, (num_data1, 1))

x2 = np.reshape(x2, (num_data2, 1))
y2 = np.reshape(y2, (num_data2, 1))
label2 = np.reshape(label2, (num_data2, 1))
```

```python
def plot_01():
    util.plot_data_label(x1, y1, label1)
```

## solve the linear regression problem by the gradient descent algorithm

`np.matmul`, `np.transpose`, `np.exp`, `np.log`

```python
k1          = 6
lr1         = 0.1
num1_iter   = 1000
```

```python
theta1      = np.zeros((k1, 1))
theta1_iter = np.zeros((k1, num1_iter))
loss1_iter  = []
t_start     = time.time()

for i in range(num1_iter):
    theta1, loss1 = util.compute_gradient_descent_step1(x1, y1, label1, theta1, lr1)
    loss1_iter.append(loss1.item())

    for j in range(k1):
        theta1_iter[j,i] = theta1[j].item()

t_end       = time.time()
t_elapsed   = t_end - t_start
```

## plot the loss over the gradient descent iterations

```python
def plot_02():
    plt.figure(figsize=(8,4))
    plt.plot(loss1_iter)
    plt.tight_layout()
    plt.show()
    print('loss1 sum =', sum(loss1_iter))
```

```python
plot_02()
```

## plot the model parameter values over the gradient descent iterations

```python
def plot_03():
    plt.figure(figsize=(8,4))
    cmap = plt.cm.jet(np.linspace(0, 1, k1))
    
    for i in range(k1):
        plt.plot(theta1_iter[i], '-', color=cmap[i])
    
    plt.xlabel('iter')
    plt.tight_layout()
    plt.show()
```

```python
plot_03()
```

## plot the classifier over the data domain

`np.arange`, `np.meshgrid`, `plt.colorbar`, `plt.contourf`, `plt.contour`, `plt.axis`, `plt.legend`, `plt.tight_layout`       

```python
def plot_04():
    util.plot_classifier1(x1, y1, label1, theta1)
```

```python
plot_04()
```

## plot the elapsed computation time for the optimization

```python
def plot_05():
    print('elapsed time =', t_elapsed)
```

```python
plot_05()
```
