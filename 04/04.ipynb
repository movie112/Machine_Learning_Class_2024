# Isotropic Diffusion via Heat Equation

## import library

```python
import numpy as np
import matplotlib.image as img
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import util
```

## load image 

```python
I_color = img.imread('04_image.jpeg')   # load image
```

## visualize the input image in color

```python
plt.figure(figsize=(8,8))
plt.imshow(I_color)
plt.title('input image in color')
plt.axis('off')
plt.show()
```

## check the size of image

```python
dim_height  = I_color.shape[0]
dim_width   = I_color.shape[1]
dim_channel = I_color.shape[2]
print('height=', dim_height, ', width=', dim_width, ', channel=', dim_channel)
```

## convert the input image into gray scale by taking the average over the channel

```python
I_gray = util.convert_color_to_gray(I_color)
```

## normalize the gray-scale image into the range [0, 1]

```python
I_norm = util.normalize_01(I_gray)
```

```python
val_min = np.min(I_norm)
val_max = np.max(I_norm)
print('min=', val_min, ', max=', val_max)
```

## visualize the input image in gray

```python
plt.figure(figsize=(8,8))
plt.imshow(I_norm, cmap='gray', vmin=0, vmax=1)
plt.title('input image in gray scale')
plt.axis('off')
plt.show()
```

## apply the heat equation on the image data with different diffusion times

```python
delta_t     = 0.1 
num_iter1   = 10
out1        = util.compute_heat_equation(I_norm, delta_t, num_iter1)
```

```python
def plot_01():
    plt.figure(figsize=(8,8))
    plt.imshow(out1, cmap='gray')
    plt.axis('off')
    plt.show()
    print('sum =', out1.sum())
```
