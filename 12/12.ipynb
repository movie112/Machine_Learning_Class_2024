# Principal Component Analysis

## import library

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
from matplotlib import cm
import time                                                
import util
```

## load data

```python
fname_data  = '12_data.csv'
data        = np.genfromtxt(fname_data, delimiter=',')
num_data    = np.size(data, 0)
num_feature = np.size(data, 1)
data        = np.reshape(data, (num_data, num_feature))
```

## plot the data

```python
plt.figure(figsize=(8,8))   
plt.scatter(data[:,0], data[:,1], color='blue')
plt.tight_layout()
plt.show()
```

## Normalization (Z-scoring)

```python
data_norm = util.normalize(data)
```

```python
def plot_01():
    plt.figure(figsize=(8,8))   
    plt.scatter(data_norm[:,0], data_norm[:,1], color='blue')
    plt.plot(0, 0, 'x', color='red')
    plt.tight_layout()
    plt.show()
```

```python
plot_01()
```

```python
def plot_02():
    print('mean = (%.2f, %.2f)' % (data_norm.mean(axis=0)[0], data_norm.mean(axis=0)[1]))
    print('std = (%.2f, %.2f)' % (data_norm.std(axis=0)[0], data_norm.std(axis=0)[1]))
```

```python
plot_02()
```

## compute principal components

`np.matmul`, `np.transpose`, `np.linalg.eig`, `np.argsort`, `plt.quiver`, `plt.scatter`

```python
pc1 = util.get_principal_component_first(data_norm)
```

```python
pc2 = util.get_principal_component_second(data_norm)
```

```python
def plot_03():
    print('first principal component = (%.2f, %.2f)' % (pc1[0], pc1[1]))
    print('second principal component = (%.2f, %.2f)' % (pc2[0], pc2[1]))
    
```

## plot the principal components

```python
plot_03()
```

```python
def plot_04():
    util.plot_principal_component(data_norm, pc1, pc2)
```

```python
plot_04()
```

## plot the projection of the data on the first principal components

```python
def plot_05():
    util.plot_projection_principal_component(data_norm, pc1)
```

```python
plot_05()
```

## plot the projection of the data on the secpmd principal components

```python
def plot_06():
    util.plot_projection_principal_component(data_norm, pc2)
```

```python
plot_06()
```
