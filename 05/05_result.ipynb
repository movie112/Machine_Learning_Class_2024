# Taylor approximation

## import library

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import util_solution as util
```

## define the first function $f(x) = \cos(x)$

```python
x1 = np.linspace(-np.pi,np.pi,100)
f1 = util.get_fun1(x1)
```

```python
def plot_01():
    plt.figure(figsize=(8,8))
    plt.plot(x1, f1)
    plt.show()
    print('sum =', f1.sum()) 
```

```python
plot_01()
```

## compute the first derivative of the first function $f(x) = \cos(x)$

```python
y2 = util.get_derivative_fun1_first(x1)
```

```python
def plot_02():
    plt.figure(figsize=(8,8))
    plt.plot(x1, y2)
    plt.show()
    print('sum =', y2.sum())
```

```python
plot_02()
```

## compute the first order Taylor approximation of the first function $f(x)=\cos(x)$ at $z=-0.5$

```python
z1 = -0.5
y3 = util.get_taylor_approximate_fun1_first(x1, z1)
```

```python
def plot_03():
    plt.figure(figsize=(8,8))
    plt.plot(x1, f1)
    plt.plot(x1, y3)
    plt.plot(z1, util.get_fun1(z1), 'o')
    plt.show()
    print('sum =', y3.sum())
```

```python
plot_03()
```
